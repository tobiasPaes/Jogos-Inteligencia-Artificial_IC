```python
import time

TAM_TABULEIRO = 3

# classe para cricao do jogo da velha e implementacao do algoritmo min max 
class Jogo:

    # inicializa o estado inicial (ainda disponivel para jogar) e jogadores
    def __init__(self, estado_inicial, jogador1, jogador2):
        self.estado_inicial = estado_inicial
        self.jogador1 = jogador1
        self.jogador2 = jogador2
        self.iniciar_jogo()

    # realiza a criacao do tabuleiro e inicia a primeira rodada
    def iniciar_jogo(self):
        tabuleiro = []
        for j in range(TAM_TABULEIRO):
            tabuleiro.append([self.estado_inicial for i in range(TAM_TABULEIRO)])
        
        self.tabuleiro = tabuleiro
        self.turno = self.jogador1

    # mostra o tabuleiro a cada rodada
    def criar_tabuleiro(self):
        for i in range(TAM_TABULEIRO):
            for j in range(TAM_TABULEIRO):
                print('{}|'.format(self.tabuleiro[i][j]), end=" ")
            print()
        print()


    # verifica se a jogada e permitida
    def verifica_movimento(self, px, py):
        if px < 0 or px > (TAM_TABULEIRO - 1) or py < 0 or py > (TAM_TABULEIRO - 1):
            return False
        elif self.tabuleiro[px][py] != self.estado_inicial:
            return False
        else:
            return True

    # verifica se o jogo terminou. Em caso positivo, retorna o resultado. Caso contrário, habilita a proxima rodada
    def verifica_resultado(self):
        
        # verifica vitoria vertical
        for i in range(TAM_TABULEIRO):
            if (self.tabuleiro[0][i] != self.estado_inicial and
                self.tabuleiro[0][i] == self.tabuleiro[1][i] and
                self.tabuleiro[0][i] == self.tabuleiro[2][i]):
                return self.tabuleiro[0][i]

        # verifica vitoria horizontal
        for i in range(TAM_TABULEIRO):
            if (self.tabuleiro[i] == [self.jogador1, self.jogador1, self.jogador1]):
                return self.jogador1
            elif (self.tabuleiro[i] == [self.jogador2, self.jogador2, self.jogador2]):
                return self.jogador2

        # verifica a diagonal principal
        if (self.tabuleiro[0][0] != self.estado_inicial and
            self.tabuleiro[0][0] == self.tabuleiro[1][1] and
            self.tabuleiro[0][0] == self.tabuleiro[2][2]):
            return self.tabuleiro[0][0]

        # verifica a diagonal secundaria
        if (self.tabuleiro[0][2] != self.estado_inicial and
            self.tabuleiro[0][2] == self.tabuleiro[1][1] and
            self.tabuleiro[0][2] == self.tabuleiro[2][0]):
            return self.tabuleiro[0][2]

        # verifica se ainda ha jogadas disponiveis
        for i in range(TAM_TABULEIRO):
            for j in range(TAM_TABULEIRO):
                if (self.tabuleiro[i][j] == self.estado_inicial):
                    return None

        # se chegou ate aqui, o resultado e empate
        return self.estado_inicial

    # implementacao do jogador max (nesse caso a ia)
    def jogador_max(self):

        # inicializao com o pior cenario possivel
        max = -2

        px = None
        py = None

        # retorna o resultado do jogo ao terminar
        resultado = self.verifica_resultado()

        if resultado == self.jogador1:
            return (-1, 0, 0)
        elif resultado == self.jogador2:
            return (1, 0, 0)
        elif resultado == self.estado_inicial:
            return (0, 0, 0)

        for i in range(TAM_TABULEIRO):
            for j in range(TAM_TABULEIRO):
                if self.tabuleiro[i][j] == self.estado_inicial:
                    
                    # jogador max faz uma jogada
                    self.tabuleiro[i][j] = self.jogador2

                    # chama min
                    (m, min_i, min_j) = self.jogador_min()
                    
                    # atualizacao do valor max
                    if m > max:
                        max = m
                        px = i
                        py = j
                    
                    # retorna o campo para o estado inicial
                    self.tabuleiro[i][j] = self.estado_inicial

        return (max, px, py)


    # implementacao do jogador min (nesse caso a pessoa)
    def jogador_min(self):

        # inicializacao do valor min com o pior cenario possivel
        min = 2

        qx = None
        qy = None

        # retorna o resultado do jogo ao terminar
        resultado = self.verifica_resultado()

        if resultado == self.jogador1:
            return (-1, 0, 0)
        elif resultado == self.jogador2:
            return (1, 0, 0)
        elif resultado == self.estado_inicial:
            return (0, 0, 0)

        for i in range(TAM_TABULEIRO):
            for j in range(TAM_TABULEIRO):
                if self.tabuleiro[i][j] == self.estado_inicial:

                    # jogador min faz uma jogada
                    self.tabuleiro[i][j] = self.jogador1

                    # chama max 
                    (m, max_i, max_j) = self.jogador_max()

                    # atualizacao do valor min
                    if m < min:
                        min = m
                        qx = i
                        qy = j

                    # retorna o campo para o estado inicial
                    self.tabuleiro[i][j] = self.estado_inicial

        return (min, qx, qy)

    # inicia o processo de busca do algoritmo min max
    def jogar(self):

        pos1 = self.tabuleiro[0][0]
        pos2 = self.tabuleiro[0][1]
        pos3 = self.tabuleiro[0][2]
        pos4 = self.tabuleiro[1][0]
        pos5 = self.tabuleiro[1][1]
        pos6 = self.tabuleiro[1][2]
        pos7 = self.tabuleiro[2][0]
        pos8 = self.tabuleiro[2][1]
        pos9 = self.tabuleiro[2][2]
        
        while True:
            self.criar_tabuleiro()
            self.resultado = self.verifica_resultado()

            # mostra o resultado ao final do jogo
            if self.resultado != None:
                if self.resultado == self.jogador1:
                    print('O vecedor é {}!'.format(self.jogador1))
                elif self.resultado == self.jogador2:
                    print('O vecedor é {}!'.format(self.jogador2))
                elif self.resultado == self.estado_inicial:
                    print("Empate!")

                self.iniciar_jogo()
                return

            # rodada do jogador 1 (humano)
            if self.turno == self.jogador1:

                while True:

                    inicio = time.time()
                    # inicia a busca pela melhor jogada
                    (m, qx, qy) = self.jogador_min()
                    fim = time.time()
                    print('Tempo para cálculo da jogada: {}s'.format(round(fim - inicio, 7)))
                    #print('Movimento recomendado: Linha = {}, Coluna = {}'.format(qx, qy))

                    jogada = int(input('Informe onde vai jogar: '))
                    
                    if jogada == 1:
                        jogada = pos1     
                        px = 0
                        py = 0 
                          

                    elif jogada == 2:
                        jogada = pos2
                        px = 0
                        py = 1
                        
                    elif jogada == 3:
                        jogada = pos3
                        px = 0
                        py = 2
                         
                    
                    elif jogada == 4:
                        jogada = pos4  
                        px = 1
                        py = 0                      
                         
                    
                    elif jogada == 5:
                        jogada = pos5
                        px = 1
                        py = 1
                        
                    
                    elif jogada == 6:
                        jogada = pos6
                        px = 1
                        py = 2
                         
                    
                    elif jogada == 7:
                        jogada = pos7 
                        px = 2
                        py = 0 
                                          
                    
                    elif jogada == 8:
                        jogada = pos8
                        px = 2
                        py = 1
                        
                    
                    elif jogada == 9:
                        jogada = pos9
                        px = 2
                        py = 2
                        
                    if self.verifica_movimento(px, py):
                            self.tabuleiro[px][py] = self.jogador1
                            self.turno = self.jogador2
                            break
                    else:
                        print('Movimento inválido. Tente novamente!')


            # rodada do jogador 2 (ia)
            else:
                (m, px, py) = self.jogador_max()
                self.tabuleiro[px][py] = self.jogador2
                self.turno = self.jogador1
```
